## Question 1 Exploratory data analysis


## 1. Import libraries

```python
import json
import numpy as np
import pandas as pd
from pprint import pprint
from pandas.io.json import json_normalize
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.tseries.holiday import USFederalHolidayCalendar, get_calendar
%matplotlib inline  
```

## 2.Import data and aggregation by time

```python
with open('./data/logins (2).json') as data_file:    
    data = json.load(data_file)
login_time = data['login_time']
print(login_time[:10]) 
```

Aggregate the timestamps by 15-minutes interval, and generate a quick plot

```python
plt.figure(figsize=(15,6))
dates = [pd.Timestamp(x) for x in  login_time]
sr = pd.Series(np.repeat(1, len(dates)), index = dates)
sr_15min = sr.resample('15T').agg('sum')
sr_15min.plot()
plt.title('15-minute demand')
plt.ylabel('15-minute demand')
plt.xlabel('15-minute interval')
```


## 3. Exploratory Analysis
### *Create new variables*
Create dataframe from the timestamps, and calculate some useful variables for exploratory analysis, like month, weekday, etc.

```python
df_15min = pd.DataFrame({'time_15min':sr_15min.index, 'count':sr_15min.values})
df_15min['weekday_name'] = df_15min.time_15min.dt.weekday_name
df_15min['weekday'] = df_15min.time_15min.dt.weekday
df_15min['month'] = df_15min.time_15min.dt.month
df_15min['date'] = df_15min.time_15min.dt.date
df_15min['week_of_year'] = df_15min.time_15min.dt.weekofyear
df_15min['day_of_year'] = df_15min.time_15min.dt.dayofyear
df_15min['hour'] = df_15min.time_15min.dt.hour
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df_15min['weekday_name'] = pd.Categorical(df_15min['weekday_name'], days)
newcal = get_calendar('USFederalHolidayCalendar')
holidays = newcal.holidays(start='2010-01-01', end='2010-12-31')
df_15min['if_holiday'] = [1 if x in holidays else 0 for x in df_15min.date]
df_15min.sample(10)
```

## Question 2 Experiment Design


## 1. Measure of success

- For a given population of drivers, I would like to choose the measure of success as:
    - the increase in number of Uber trips that drivers started in the other city (not their home city) after toll imbursement


- The reason is that if drivers are incentivized by the toll imbursement, they are more likely to cross the bridge to pick up passengers because they don't need to pay the toll themselves. Since there might be a supply-demand imbalance in weekdays for the two cities, some passengers whose demand was not met previously may now have drivers willing to pick them up from the city on the other side of the bridge. 


- After crossing the bridge, the driver may choose to stay in the other city to pick up more passengers. These could be reflected by the increase in number of Uber trips started by drivers coming from the other city.

## 2. Experiment Design



### * (a) How I will implement the experiment*
- Identify the total population of Uber drivers serving the market of these two cities (i.e., drivers registered in Gotham or Metropolis) at the beginning of the experiment. For simplicity, I would assume there is a total of 1000 drivers.


- The idea is to divide the driver group into control group and treatment group, and study the impact of toll reimbursement. However, there might be other important confounding factors that has an impact on the casual relationship. here I assume that drivers may differ in their reaction to the toll reimbursement depending on whether they has previously crossed the bridge to pick up passengers in other city or not. There may be other factors that can be useful to divide the group.


- Assume that of the total population of 1000 drivers, there is a total of 300 drivers who have crossed the bridge before to pick up passengers. The population can be randomly divided into 4 sub-groups.

||Control (**C**)|Treatment (**T**)|
|------|------|------|
|Has previous cross-bridge pick-ups (**Y**)|150|150|
|Has no previous cross-bridge pick-ups (**N**)|350|350|

- For the 500 drivers selected in the treatment group, Uber could send notification to their phones and inform them of the toll imbursement policy. Their tolls for using the bridge will be reimbursed after finishing the cross-bridge trip.

### * (b) Statistical test*

#### **Confirm Pre-experiment similarity**

- A two-factor ANOVA (Analysis of variance) model can be used to identify the impact of treatment (toll reimbursement), drivers' previous cross-bridge pick-ups and their interaction. 


- Before comparing the post-experiment trend for trips starting in the other city, It is important to ensure that the two groups of drivers (control and treatment) have the same trend before the experiment starts. After confirming that they have the same pre-experiment pattern, we can be more confident that the difference after the experiment is due to the treatment, not the existing difference between two groups.


- I would collect the number of historical daily trips starting from the other side of the bridge by all drivers in the control and treatment group. I assume the pre-experiment period lasts one month (e.g., from July 1 to July 31). A **paired t-test** can be used to verify that the two groups are not different statistically with the data collected as below.

- Note that since there might be a difference between weekend and weekday demand pattern, the ANOVA table can be calculated for weekday and weekend data, respectively. In the table below, $\mathbf{X}^{CY}$ = [$X_{day1}^{CY}$, $X_{day2}^{CY}$, ......], similar for other cells

### * (c) Result interpretation*


- It is expected that toll reimbursement would increase trips in the other city. 
    - $\bar{\mathbf{X}}^{TY}$ > $\bar{\mathbf{X}}^{CY}$
    - $\bar{\mathbf{X}}^{TN}$ > $\bar{\mathbf{X}}^{CN}$
    
    
- It is expected that $P_1$<0.05 in the AVONA table, which indicates that under a significance level of 0.05, toll reimbursement has a significant impact on number of trips in the other city.


- It might be expected that $P_3$<0.05. Intuitively, toll reimbursement should be more effective in encouraging drivers who have not made cross-bridge pick-ups before to start serve trips in the other city. For drivers who have already made cross-bridge pick-ups, maybe they are already getting used to paying the toll by themselves, so toll reimbursement made little impact to these insensitive drivers.


- It is possible that the significant difference is only demonstrated on weekdays instead of weekends

### *Possible recommendation*
- Assuming that the statistical significances are consistent with expectation, the city operation team may want to convert the change in number of trips to the increase in revenue. Together with the cost of toll reimbursement for drivers, this would enable the operation team to conduct a benefit-cost analysis to find out if the growing number of trips in the other city would really increase revenue for Uber (short-term and long-term).


- Based on the result from the cost-benefit analysis, the operation team may have the following possible options
    - Do not continue toll reimbursement because few drivers are impacted and the overall benefit is not enough to justify toll reimbursement 
    - Only offer toll reimbursement to drivers without previous experience of cross-bridge pick-ups.
    - Only offer toll reimbursement when and where is a supply-demand imbalance (e.g.,weekdays).

## Question 3 Predictive modeling

## 1. Cleaning, visualization and exploratory analysis

### *Import addtional libraries*

```python
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import metrics
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.metrics import confusion_matrix,accuracy_score, roc_curve,auc
```

### *Import data*

```python
with open('./data/MODS_DS_data_challenge.json') as data_file:
    data = json.load(data_file)   
df = json_normalize(data)
df.sample(10)
```

### *Clean data formats and create variables for date/time*

```python
df['signup_date'] = pd.to_datetime(df.signup_date)
df['last_trip_date'] = pd.to_datetime(df.last_trip_date)
df['signup_weekday_name'] = df.signup_date.dt.weekday_name
df['signup_weekday'] = df.signup_date.dt.weekday
df['signup_month'] = df.signup_date.dt.month
df['signup_day_of_year'] = df.signup_date.dt.dayofyear
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df['signup_weekday_name'] = pd.Categorical(df['signup_weekday_name'], days)
```

### *Calculate fraction of active users*
All the users in the dataset are signed up in January, 2014. Therefore, if a user's last trip date is later than June 1, 2014, it indicates that this user is defined as an active user. There is a total of **37.6%** active users in the dataset.

```python
df['active'] = 0
df.loc[df.last_trip_date >= '2014-06-01','active'] = 1
percent_active = sum(df.active) / len(df)
print('The percent of active user is {:.1%}'.format(percent_active))
```

### *Split the train and test data*
30% is randomly selected as testing dataset, and the rest 70% is used as training dataset to develop model.

```python
y = df.active
X = df.drop('active', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=888)
```

### *Data type and missing values*
Most of the columns has no missing value or a very low (<1%) proportion of missing value. However, it should be noted that **`avg_rating_of_driver`** has a 16% percent of missing value.

```python
print('----The training set has a total of {0} records----\n'.format(len(X_train)))
print('----Perent of missing values----')
X_train.isnull().sum()/len(X_train)*100
```

Since **`avg_rating_of_driver`** has a high percent of missing values, and the value missing may be related to the dependent variable itself, it is not a good idea to simply impute it with average/median. An additional dummy variable is added
- **`has_rating_of_driver`**: (0: avg_rating_of_driver is NA, otherwise 1)


```python
X_train = X_train.assign(has_rating_of_driver=([0] * len(X_train)))
X_train.loc[X_train.avg_rating_of_driver.notnull(), 'has_rating_of_driver'] = 1
```

### *Define categorical and continuous columns*

```python
categ_cols = [
    'city',
    'phone',
    'uber_black_user',
    'signup_weekday_name',
    'signup_day_of_year',
    'has_rating_of_driver']

conti_cols = [
 'avg_dist',
 'avg_rating_by_driver',
 'avg_rating_of_driver',
 'avg_surge',
 'surge_pct',
 'trips_in_first_30_days',
 'weekday_pct']

print('There are totally {0} categorical variables and {1} continous variable'.format(len(categ_cols), len(conti_cols)))
```

Since the missing value rate is very low for other variables, I just imputed missing values for categorical and continuous values.
- Impute categorical variable NA with the most frequent item
- Impute continuous NA with median

```python
for column in categ_cols:
    X_train[column].fillna(X_train[column].mode()[0],inplace=True)
for column in conti_cols:
    X_train[column].fillna(np.nanmedian(X_train[column]),inplace=True)
```

### *Exploratory Analysis for Categorical Variables*

```python
df = pd.concat([X_train, y_train], axis = 1) #get df with X and y for plotting purpose
```

The figure below shows the **number of records** categorized by categorical variables. 
- The majority of users are iphone users
- The number of sign-ups on Fridays and Saturdays are higher than the rest of the week

```python
plt.figure(figsize=(15,12))
sns.set_context("notebook",font_scale=1.25)
fig_no = 1
for column in categ_cols:
    plt.subplot(3,2,fig_no)
    sns.barplot(x=column, y='active', data=df, estimator = len, ci = None, color='seagreen')
    fig_no+=1
    plt.ylabel('Number of records')
```

The figure below shows the **percent of active riders** categorized by some variables. 
- The city where the user signed up seems to impact the percent of active users, with King's Landing has highest percent of active riders
- Iphone users are more likely to be active riders compared with Android users
- Uber black users are more likely to be active riders

### 2. Model development

The following evaluation metrics will be used:
- Accuracy
- Confusion Matrix
- ROC curve and AUC (Area Under Curve)

It should be noted that in our training set, there are around one-third of active users, so the classes are not very well-balanced. Just using accuracy fails to account for the class imbalance. Confusion matrix will provide useful information on True/False Positive/Negative. AUC describes the probability that a randomly chosen positive example ranks above a randomly chosen negative example. ROC curve and AUC value will be a better option to select between models and tune model parameters.

### *Repeat the feature engineering for test dataset*

```python
categ_cols.remove('signup_day_of_year') # this variable is just used for exploratory analysis
feature = conti_cols + categ_cols
X_train = X_train[feature]

X_test = X_test.assign(has_rating_of_driver=([0] * len(X_test)))
X_test.loc[X_test.avg_rating_of_driver.notnull(), 'has_rating_of_driver'] = 1

for column in categ_cols:
    X_test[column].fillna(X_train[column].mode()[0],inplace=True)
for column in conti_cols:
    X_test[column].fillna(np.nanmedian(X_train[column]),inplace=True)

X_test = X_test[feature]
```

### *Code dummy variables for categorical variables*

Categorical variable with L possible values are coded with (L-1) dummy variables.

```python
for column in categ_cols:
    temp=pd.get_dummies(X_train[column], drop_first=True, prefix = column)
    X_train.drop(column, axis = 1, inplace=True)
    X_train=pd.concat([X_train,temp], axis=1)
    
for column in categ_cols:
    temp=pd.get_dummies(X_test[column], drop_first=True, prefix = column)
    X_test.drop(column, axis = 1, inplace=True)
    X_test=pd.concat([X_test,temp], axis=1)    
```

### Logistic Regression
I would like to start with logistic regression:
- Logistic regression is a typical type of regression having a binary response variable.
- Feature selection with Lasso can be combined with logistic regression to avoid over-fitting
- The estimated coefficients are more interpretable compared with other methods.


The lasso regularization imposes penalty with same weights on the coefficients, so the variables need to be standardized for training. **`LogisticRegressionCV`** is used to find the best regularization parameter by implementing a grid search for C using 5-fold cross-validation. AUC value is used as the score to select best parameter.

```python
lr_cv = LogisticRegressionCV(penalty='l1', solver='liblinear', 
                             Cs=[100, 10, 1, 0.1, 0.01, 0.001],class_weight = 'balanced',
                             cv = 5,scoring = 'roc_auc',refit=True)
lr_cv.fit(X = X_train / np.std(X_train, 0), y = y_train)
print('The best C is '+ str(lr_cv.C_[0]))
```

An increasing C value indicates less penalty for coefficients. The best C value is 0.1 from gridCV. Personally I prefer the least complicated model with reasonable performance. As shown in the figure below, C= 0.01 gives similar performance compared with larger C values while giving highest penalty for coefficients, so I am using C=0.01 for final model.

```python
lr_cv_scores = pd.DataFrame(lr_cv.scores_[1]).T
lr_cv_scores.index = [100, 10, 1, 0.1, 0.01, 0.001]
lr_cv_scores = lr_cv_scores.stack().reset_index().\
            rename(columns={'level_0': 'C', 'level_1': 'Fold',0: 'auc_roc'})
sns.pointplot(x='C', y='auc_roc', data = lr_cv_scores, estimator = np.mean, ci = 95)
```

The model is fitted using selected parameters. The table below shows the estimated coefficients. Most of the signs for coefficients are consistent with the observations from exploratory analysis. The top five variables are:
- `city_King's Landing`
- `phone_iPhone`
- `trips_in_first_30_days`
- `uber_black_user_True`
- `has_rating_of_driver_1`

```python
lr_ =LogisticRegression(penalty='l1', solver='liblinear', C=0.01, class_weight = 'balanced')
lr_ =lr_.fit(X = X_train / np.std(X_train, 0), y = y_train)

print('The intercept is {0:2.3f}'.format(lr_.intercept_[0]))
coef_df = pd.DataFrame({'variable':X_train.columns, 'coefficient':lr_.coef_[0]})
coef_df['sort'] = coef_df.coefficient.abs()
coef_df.sort_values(by = 'sort',ascending=0).drop('sort', axis=1)
```

### *Apply model to predict test dataset*

```python
X_test_lr = X_test / np.std(X_test, 0)
y_hat_lr = lr_.predict_proba(X = X_test_lr)[:,1]
preds_lr = lr_.predict(X = X_test_lr)
CF = pd.DataFrame(confusion_matrix(y_test,preds_lr))
CF = CF / len(X_test)
print(CF)
print('\nAccuracy: {0:2.3f}'.format(accuracy_score(y_test, preds_lr,normalize = True)))
```

The resuls indicates that the prediction for test dataset has an accuracy of **69.7%**. False Negative and False Positive rates are **11%** and **19%**, respectively. The AUC value is **0.77**.

```python
def ROC_plot(y_test, y_hat):
    fpr, tpr, _ = metrics.roc_curve(y_test, y_hat)
    roc_auc = auc(fpr, tpr)
    df = pd.DataFrame(dict(fpr=fpr, tpr=tpr))
    sns.set_context("notebook", font_scale=1.2)
    plt.figure()
    plt.plot(df.fpr, df.tpr, color='darkorange',
             lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")

ROC_plot(y_test, y_hat_lr)   
```

### Gradient Boosting Trees

One main shortcoming of logistic regression is that it only considered the linear relationship between independent variable and the outcome (i.e., the decision boundary is linear). Decision tree captures the non-linear relationship by nature, and boosting algorithm would reduce the variance of decision tree by sequentially building weak classifiers, and aggregating them.

Parameter tuning need to be done before getting a fitted model. Grid search is again used to find the best value for the following parameters. A more complete parameter tuning can include other paramters as well. Here I have done a quick tuning for the parameters below.
- Number of estimators
- The maximum depth of the tree.
- The minimum number of samples required to split an internal node

```python
param_test1 = {'n_estimators':range(10,101,10)}
gsearch1 = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1,min_samples_split=500,min_samples_leaf=50,max_depth=5,max_features='sqrt',subsample=0.8,random_state=10), 
param_grid = param_test1, scoring='roc_auc',cv=5,n_jobs=4)
gsearch1.fit(X_train,y_train)
```

```python
gsearch1.cv_results_['mean_test_score'], gsearch1.cv_results_['std_test_score'], gsearch1.best_params_, gsearch1.best_score_
```

```python
param_test2 = {'max_depth':range(5,10,2), 'min_samples_split':range(500,1001,200)}
gsearch2 = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1, 
min_samples_leaf=50,max_features='sqrt',subsample=0.8,random_state=10, n_estimators=100),  
param_grid = param_test2, scoring='roc_auc',cv=5,n_jobs=4)
gsearch2.fit(X_train,y_train)
```

```python
gsearch2.cv_results_['params'],gsearch2.cv_results_['mean_test_score'], gsearch2.cv_results_['std_test_score'], gsearch2.best_params_, gsearch2.best_score_
```

The model is fitted using the best parameter set:

```python
gb_ = GradientBoostingClassifier(learning_rate=0.1, min_samples_leaf=50,max_depth=9, min_samples_split= 900,max_features='sqrt',subsample=0.8,random_state=10, n_estimators=100)
gb_.fit(X_train, y_train)
```

Feature importance is summarized and plotted

Apply model to predict test dataset.

```python
y_hat_gb = gb_.predict_proba(X = X_test)[:, 1]
preds_gb = gb_.predict(X = X_test)
CF = confusion_matrix(y_test,preds_gb)
CF = CF / len(X_test)
print(CF)
print('\nAccuracy: {0:2.3f}'.format(accuracy_score(y_test, preds_gb,normalize = True)))
```

The result indicates that the prediction for test dataset has an accuracy of **79.1%**. False Negative and False Positive rates are **12%** and **9%**, respectively. The AUC value is **0.86**.
